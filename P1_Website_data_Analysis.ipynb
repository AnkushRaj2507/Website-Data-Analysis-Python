# WEBSITE DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ankus\OneDrive\Desktop\SDE+DATA\DATA\Projects\Python Website Data analysis\data-export (1).csv")
df
```

```python
df.columns= df.iloc[0]
df=df.drop(index=0).reset_index(drop=True)
df.columns= ['Channel Group', 'Date Hour','Users', 'Sessions', 'Engaged sessions', 'Average engagement time per session', 'Engaged sessions per user',	'Events per session',	'Engagement rate', 'Event Count' ]
df
```

```python
df.info()
```

```python
df['Date Hour']= pd.to_datetime(df['Date Hour'], format='%Y%m%d%H', errors='coerce')

numeric_columns = df.columns.drop(['Channel Group', 'Date Hour'])
df[numeric_columns]= df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df['Hour']= df['Date Hour'].dt.hour

df
```

```python
df.describe()
```

# Insights:

# 1. Website sessions and user over time

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6), dpi=150)
df.groupby('Date Hour')[['Sessions', 'Users']].sum().plot()
plt.title('Website Sessions and Users Over Time')
plt.xlabel('Date Hour')
plt.ylabel('Count')
plt.show()
```

# 2. Marketing Channel and Number of Users

```python

plt.figure(figsize=(10,3), dpi=150)
sns.barplot(x='Channel Group', y='Users', data=df, estimator=sum , palette="viridis" )
plt.title('Marketing Channel and Number of Users')
plt.xlabel('Marketing Channel')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()
```

# 3. Marketing Channel and Average Engagement Time

```python
plt.figure(figsize=(8,3), dpi=150)
sns.barplot(x='Channel Group', y='Average engagement time per session', data=df, estimator=np.mean , palette="magma" )
plt.title('Marketing Channel and Average Engagement Time')
plt.xlabel('Marketing Channel')
plt.ylabel('Average engagement time per session')
plt.xticks(rotation=45)
plt.show()
```

# 4. Engagement Rate across Different traffic Channels

```python
plt.figure(figsize=(8,3), dpi=150)
sns.boxplot(x='Channel Group', y='Engagement rate', data=df , palette="coolwarm" )
plt.title('Marketing Channel and Average Engagement Time')
plt.xlabel('Marketing Channel')
plt.ylabel('Engagement rate')
plt.xticks(rotation=45)
plt.show()
```

# 5. Engaged Sessions vs Non Engaged Sessions

```python
session_df= df.groupby('Channel Group')[['Sessions', 'Engaged sessions']].sum().reset_index()
session_df['Non Engaged sessions']= session_df['Sessions'] - session_df['Engaged sessions']
session_df= session_df.melt(id_vars='Channel Group', value_vars=['Engaged sessions', 'Non Engaged sessions'])


plt.figure(figsize=(8,3), dpi=150)
sns.barplot(x='Channel Group', y='value', data=session_df ,hue="variable", palette="Set2" )
plt.title('Engaged Sessions vs Non Engaged Sessions')
plt.xlabel('Marketing Channel')
# plt.ylabel('Engaged sessions')
plt.xticks(rotation=45)
plt.show()
```

# 6. Traffic Hours of the Day

```python
heat_map_data= df.groupby(['Hour', 'Channel Group'])['Sessions'].sum().unstack().fillna(0)

plt.figure(figsize=(12,6), dpi=150)
sns.heatmap(heat_map_data, cmap='YlGnBu',linewidths=0.5, annot=True, fmt=".0f")
plt.title('Marketing Channel and Average Engagement Time')
plt.xlabel('Channel Group')
plt.ylabel('Hour of the Day')
plt.xticks(rotation=45)
plt.show()
```

# 7. High Engagemengt Rate vs Sessions over time

```python
df_plot= df.groupby('Date Hour')[['Sessions', 'Engagement rate']].mean().reset_index()

plt.figure(figsize=(12,6), dpi=150)
plt.plot(df_plot['Date Hour'], df_plot['Sessions'], label='Sessions', color='orange')
plt.plot(df_plot['Date Hour'], df_plot['Engagement rate'], label='Engagement rate', color='red')
plt.title('High Engagemengt Rate vs Sessions over time')
plt.xlabel('Date Hour')
plt.ylabel('High Engagemengt Rate and Sessions')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
